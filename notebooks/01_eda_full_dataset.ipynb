```python
import numpy as np
import pandas as pd
from pathlib import Path

# المسار إلى جذر المشروع
BASE_DIR = Path(r"C:\Users\islem\Desktop\R37-LightGBM")

RAW_DATA_DIR = BASE_DIR / "data" / "raw"
PROC_DATA_DIR = BASE_DIR / "data" / "processed"
PROC_DATA_DIR.mkdir(exist_ok=True)

EDP_PATH = RAW_DATA_DIR / "edp.csv"

print("BASE_DIR     :", BASE_DIR)
print("RAW_DATA_DIR :", RAW_DATA_DIR)
print("EDP_PATH     :", EDP_PATH)
```

```python
edp = pd.read_csv(EDP_PATH)

print("Shape:", edp.shape)
print("\nColumns (53 expected):")
print(edp.columns.tolist())

print("\nHEAD:")
display(edp.head(3))

print("\nINFO:")
print(edp.info())

print("\nDESCRIBE (numeric):")
display(edp.describe().T)
```

```python
# 1) كل أعمدة SEGMENT_2 كـ features (36 عمودًا)
feature_cols = [c for c in edp.columns if c.startswith("SEGMENT_2")]

print("=== FEATURES ===")
print("Number of features:", len(feature_cols))
print(feature_cols)

# 2) اختيار الأهداف (targets) – نفس الـ6 السابقة كبداية
target_cols = [
    "computation_pout_C_ISEN_EFF_TOT2TOT",
    "computation_pout_C_POLY_EFF_TOT2TOT",
    "computation_pout_TOT_PRES_RATIO",
    "computation_pout_MASS_FLOW",
    "computation_pout_TORQUE",
    "computation_pout_C_PRES_LOSS",
]

print("\n=== TARGETS ===")
print("Number of targets:", len(target_cols))
print(target_cols)
```

```python
X = edp[feature_cols].values
y = edp[target_cols].values

print("X shape:", X.shape)   # (2000, n_features)
print("y shape:", y.shape)   # (2000, 6)

OUT_PATH = PROC_DATA_DIR / "processed_data_multi.npz"

np.savez(
    OUT_PATH,
    X=X,
    y=y,
    feature_cols=np.array(feature_cols, dtype=str),
    target_cols=np.array(target_cols, dtype=str),
)

print(f"\n✅ Saved processed arrays to: {OUT_PATH}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# دمج features + targets في DataFrame واحد
df_corr = edp[feature_cols + target_cols]

corr = df_corr.corr(numeric_only=True)

plt.figure(figsize=(12, 10))
sns.heatmap(corr[target_cols].loc[feature_cols],
            cmap="coolwarm", center=0)
plt.title("Correlation between SEGMENT_2 features and targets")
plt.tight_layout()
plt.show()
```
