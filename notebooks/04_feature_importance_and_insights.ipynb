```python
import numpy as np
import pandas as pd
from pathlib import Path
import joblib

BASE_DIR      = Path(r"C:\Users\islem\Desktop\R37-LightGBM")
PROC_DATA_DIR = BASE_DIR / "data" / "processed"
MODELS_DIR    = BASE_DIR / "models" / "lgbm_baseline"

DATA_PATH  = PROC_DATA_DIR / "processed_data_multi.npz"
MODEL_PATH = MODELS_DIR / "model.pkl"

data = np.load(DATA_PATH, allow_pickle=True)
X = data["X"]; y = data["y"]
feature_cols = data["feature_cols"]; target_cols = data["target_cols"]

model = joblib.load(MODEL_PATH)
print("Loaded model and data.")
```

```python
all_importances = []

for target_name, est in zip(target_cols, model.estimators_):
    imp = est.feature_importances_
    df = pd.DataFrame({
        "feature": feature_cols,
        "importance": imp,
    })
    df["target"] = target_name
    df.sort_values("importance", ascending=False, inplace=True)
    all_importances.append(df)

imp_df = pd.concat(all_importances, ignore_index=True)
```

```python
for t in target_cols:
    print(f"\n=== Top 10 features for {t} ===")
    display(imp_df[imp_df["target"] == t].head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

target_focus = "computation_pout_C_ISEN_EFF_TOT2TOT"

df_focus = imp_df[imp_df["target"] == target_focus].head(15)

plt.figure(figsize=(8, 6))
sns.barplot(data=df_focus, x="importance", y="feature")
plt.title(f"Feature importance for {target_focus}")
plt.tight_layout()
plt.show()
```
